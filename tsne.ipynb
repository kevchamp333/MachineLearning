```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

# MNIST 데이터 불러오기
data = load_digits()

# 2차원으로 차원 축소
n_components = 2

# t-sne 모델 생성
model = TSNE(n_components=n_components)

# 학습한 결과 2차원 공간 값 출력
print(model.fit_transform(data.data))
# [
#     [67.38322, -1.9517338],
#     [-11.936052, -8.906425],
#     ...
#     [-10.278599, 8.832907],
#     [25.714725, 11.745557],
# ]
tsne_np = model.fit_transform(data.data)
```

```python
import pandas as pd
```

```python
tsne_df = pd.DataFrame(tsne_np, columns = ['component_0', 'component_1'])
```

```python
tsne_df.iloc[:, 0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('digit data t-SNE result', fontsize = 14)

plt.scatter(tsne_df.iloc[:, 0], tsne_df.iloc[:, 1], color = 'pink')

plt.xlabel('component 0')
plt.ylabel('component 1')

plt.show()
```
